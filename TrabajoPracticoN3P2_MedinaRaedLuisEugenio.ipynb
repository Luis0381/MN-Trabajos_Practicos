```python
import numpy as np

def jacobi(A,n,x,b,error,maxIter): ## n: tamaño matriz - X: solucion inicial
    k=0
    numeroMayor=0
    errorCalc=error*2
    resultado=np.copy(x)
    while(k<=maxIter and errorCalc>error):
        i=0
        numeroMayor=np.max(resultado)
        while(i<n):
            j=0
            s=0
            while(j<n):
                if(j==i):
                    j=j+1
                else:
                    s=s+A[i,j]*x[j]
                    j=j+1
            resultado[i]=(b[i]-s)/A[i,i]
            i=i+1
        k=k+1    
        errorCalc=np.abs(numeroMayor-np.max(resultado))
        x=np.copy(resultado)

    print("Numero de iteraciones:")
    print(k+1)   
    print("Resultado:")
    print(x)

def gaussSeidel(A,n,x,b,error,maxIter): ## n: tamaño matriz - X: solucion inicial
    k=0
    numeroMayor=0
    errorCalc=error*2
    while(k<=maxIter and errorCalc>error):
        i=0
        numeroMayor=np.max(x)
        while(i<n):
            j=0
            s=0
            while(j<n):
                if(j==i):
                    j=j+1
                else:
                    s=s+A[i,j]*x[j]
                    j=j+1
            x[i]=(b[i]-s)/A[i,i]
            i=i+1
        k=k+1    
        errorCalc=np.abs(numeroMayor-np.max(x))

    print("Numero de iteraciones:")
    print(k+1)   
    print("Resultado:")
    print(x)
```

```python
x=np.array([1.2,1.5,1.6])

A=np.array([ [4,1,-1],
    [2,5,0],
    [-2,3,10]])

b=np.array([7,19,45])

print("Ahora resolveremos el sistema del punto 2, aplicando Jacobi y Gauss Seidel")
print("\nJacobi")
jacobi(A,3,x,b,0.001,100)
print("\n\nGauss Seidel")
gaussSeidel(A,3,x,b,0.001,100)
```
