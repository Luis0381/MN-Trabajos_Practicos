```python
## Trabajo Practico N4 - Ejercicio 4


import scipy.interpolate as sc
import numpy as np
import matplotlib.pyplot as plt


x=np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
y=np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])

cs=sc.CubicSpline(x,y) #Como entrada a la funcion se ponen los xi y f(i)

print("La grafica resultante es: \n")
xs=np.linspace(0.9, 13.3, 100) ##500 numeros igualmente espaciados entre 0.9 y 13.3
plt.plot(xs,cs(xs))
plt.grid()
plt.title('Ejercicio 4 - Trabajo Practico N4')
plt.xlabel('x')
plt.ylabel('y', fontsize=10)
plt.show()
```
